```python
%matplotlib inline
import os

import matplotlib.pyplot as plt
from pandas import read_csv
from numpy import histogram
```

```python
runs = read_csv(os.getcwd() + '/temp_cleaned.dat', sep=" ")
```

```python
print("Indices: {}".format(runs.columns.values))
print("Number of runs: {}".format(len(runs)))
```

```python
plt.xlabel("Number Steps in Run")
plt.ylabel("Number of Runs")
runs['NumberSteps'].hist(bins=50)
```

```python
plt.xlabel("Cumulative Score in Run")
plt.ylabel("Number of Runs")
runs['CumulScore'].hist(bins=50)
```

```python
plt.xlabel("Number of Steps in Run")
plt.ylabel("Cumulative Score in Run")
plt.scatter(runs['NumberSteps'], runs['CumulScore'])
```
